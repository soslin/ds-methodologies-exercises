```python
data = response.json()
pprint(data)
```

```python
sales = pd.DataFrame(data['payload']['sales'])
sales.head()
```

Print second page

```python
# response = requests.get('https://python.zach.lol' + data['payload']['next_page'])
# data = response.json()
```

```python
for i in range(1, data['payload']['max_page']):
    response = requests.get('https://python.zach.lol' + data['payload']['next_page'])
    data = response.json()
    sales = pd.concat([sales, pd.DataFrame(data['payload']['sales'])])
#sales = sales.reset_index()
```

#### Combine the data from your three separate dataframes into one large dataframe.

```python
items.shape
```

```python
items.item_id.nunique()
```

```python
stores.head()
```

```python
stores.store_id.nunique()
```

```python
sales.item.nunique()
```

```python
sales.store.nunique()
```

```python
merge_part = pd.merge(sales, items, left_on = 'item', right_on = 'item_id', how = 'inner')
```

```python
merge_total = pd.merge(merge_part, stores, left_on = 'store', right_on = 'store_id', how = 'inner')
merge_total.head()
```

```python
merge_total.to_csv('oslin_merge_total.csv')
```
